```python
from sklearn.model_selection import train_test_split
```

```python
cr_train,cr_test=train_test_split(cr,test_size=0.2,random_state=53)
cr_train_x=cr_train.iloc[:,0:-1]
cr_train_y=cr_train.iloc[:,-1]
cr_test_x=cr_test.iloc[:,0:-1]
cr_test_y=cr_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(cr_train_x,cr_train_y)
```

```python
pred_cr=logreg.predict(cr_test_x)
```

```python
pred_cr
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tab_cr=confusion_matrix(cr_test_y,pred_cr)
```

```python
tab_cr
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
```

```python
lr_acc=accuracy_score(cr_test_y,pred_cr)
lr_acc
```

```python
lr_rec=recall_score(cr_test_y,pred_cr)
lr_rec
```

```python
lr_prec=precision_score(cr_test_y,pred_cr)
lr_prec
```

```python
lr_f1=f1_score(cr_test_y,pred_cr)
lr_f1
```

```python


metrics = ['Accuracy', 'F1 Score', 'Recall', 'Precision']
scores = [lr_acc,lr_f1, lr_rec, lr_prec]
colors = ['blue', 'yellow', 'green', 'cyan']

plt.figure(figsize=(6, 4))
plt.bar(metrics, scores, color=colors, edgecolor='black')


plt.ylim(0.5, 1.0)
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Logistic Regression Performance')
plt.grid(axis='y')

# Add value labels inside or above bars based on their height
for i, score in enumerate(scores):
    label_y = score - 0.015 if score > 0.78 else score + 0.01
    va = 'top' if score > 0.78 else 'bottom'
    plt.text(i, label_y, f'{score:.2f}', ha='center', va=va, fontsize=10)

plt.tight_layout()
plt.show()
```
