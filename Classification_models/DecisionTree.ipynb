```python
from sklearn.model_selection import train_test_split
```

```python
cr_train,cr_test=train_test_split(cr,test_size=0.2,random_state=53)
cr_train_x=cr_train.iloc[:,0:-1]
cr_train_y=cr_train.iloc[:,-1]
cr_test_x=cr_test.iloc[:,0:-1]
cr_test_y=cr_test.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=5)
```

```python
dt=DecisionTreeClassifier(criterion='entropy')
```

```python
dt.fit(cr_train_x,cr_train_y)
```

```python
pred_dt=dt.predict(cr_test_x)
pred_dt
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix,classification_report
```

```python
tab_dt=confusion_matrix(cr_test_y,pred_dt)
```

```python
tab_dt
```

```python
dt_acc=accuracy_score(cr_test_y,pred_dt)
dt_acc
```

```python
dt_rec=recall_score(cr_test_y,pred_dt)
dt_rec
```

```python
dt_prec=precision_score(cr_test_y,pred_dt)
dt_prec
```

```python
dt_f1=f1_score(cr_test_y,pred_dt)
dt_f1
```

```python
print(classification_report(cr_test_y,pred_dt))
```

```python
tab_dt.diagonal().sum()/tab_dt.sum()
```

```python
import matplotlib.pyplot as plt


metrics = ['Accuracy', 'F1 Score', 'Recall', 'Precision']
scores = [dt_acc,dt_f1, dt_rec, dt_prec]
colors = ['blue', 'yellow', 'green', 'cyan']

plt.figure(figsize=(6, 4))
plt.bar(metrics, scores, color=colors, edgecolor='black')

plt.ylim(0.5, 0.85)
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Decision Tree Performance')
plt.grid(axis='y')

# Add value labels inside or above bars based on their height
for i, score in enumerate(scores):
    label_y = score - 0.015 if score > 0.78 else score + 0.01
    va = 'top' if score > 0.78 else 'bottom'
    plt.text(i, label_y, f'{score:.2f}', ha='center', va=va, fontsize=10)

plt.tight_layout()
plt.show()
```
